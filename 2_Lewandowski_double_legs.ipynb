```python
path = "/content/drive/MyDrive/Colab Notebooks/Football_Analytics/Learning/wyscout/players.json"

#open dataset with players
with open(path) as f:
    players = json.load(f)
player_df = pd.DataFrame(players)
```

```python
player_df[player_df.firstName == 'Robert']
```

```python
#take shots only
shots = df_events.loc[df_events['subEventName'] == 'Shot']
#look for Lewandowski id
RL9_id = player_df.loc[player_df["shortName"] == "R. Lewandowski"]["wyId"].iloc[0]
#get Lewandowski shot
RL9_shots = shots.loc[shots["playerId"] == RL9_id]
```

```python
#left leg shots
lefty_shots = RL9_shots.loc[RL9_shots.apply (lambda x:{'id':401} in x.tags, axis = 1)]
lefty_shots.head()
```

```python
#right leg shots
righty_shots = RL9_shots.loc[RL9_shots.apply (lambda x:{'id':402} in x.tags, axis = 1)]
righty_shots.head()
```

```python

#create list with ones for left foot shots and -1 for right foot shots
l = [1] * len(lefty_shots)
l.extend([-1] * len(righty_shots))

print(l)
```

```python
from statsmodels.stats.descriptivestats import sign_test
test = sign_test(l, mu0 = 0)
pvalue = test[1]

if pvalue < 0.05:
    print("P-value amounts to", str(pvalue)[:5], "- We reject null hypothesis - Lewadnowski is not ambidextrous")
else:
    print("P-value amounts to", str(pvalue)[:5], " - We do not reject null hypothesis - Lewandowski is ambidextrous")
```
